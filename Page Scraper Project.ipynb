```python
import requests
import bs4


def get_html(url):
    r = requests.get(url)
    return r.text

def get_all_links(html):
    soup = bs4.BeautifulSoup(html, 'lxml')
    
    links = soup.select("a", class_='cross-link')
    list_of_content = []
    
    for link in links:
        link = link.text
        list_of_content.append(link)
        
    return list_of_content
    
    

```

```python
a = get_html("https://akipress.org/")
```

```python
get_all_links(a)
```
